```sql
CREATE OR REPLACE FUNCTION kraft_udfs.domain_1.hh_inc_tier(HOUSEHOLD_INCOME_TIER INT)
RETURNS STRING
LANGUAGE SQL
RETURN
    CASE 
        WHEN HOUSEHOLD_INCOME_TIER = 1 THEN 'Budget'
        WHEN HOUSEHOLD_INCOME_TIER = 2 THEN 'Lower-Middle'
        WHEN HOUSEHOLD_INCOME_TIER = 3 THEN 'Middle'
        WHEN HOUSEHOLD_INCOME_TIER = 4 THEN 'Upper-Middle'
        ELSE 'Affluent'
    END

  
```
