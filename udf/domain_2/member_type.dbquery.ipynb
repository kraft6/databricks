```sql
CREATE OR REPLACE FUNCTION kraft_udfs.domain_2.member_type(MEMBER_TYPE INT)
RETURNS STRING
LANGUAGE SQL
RETURN
    CASE 
        WHEN MEMBER_TYPE = 1 THEN 'Primary Member'
        WHEN MEMBER_TYPE = 2 THEN 'Spouse/Partner'
        WHEN MEMBER_TYPE = 3 THEN 'Dependent'
        ELSE 'Extended Household Member'
    END
```
